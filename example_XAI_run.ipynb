### EigenCAM


```python
from XAI.cam_utils import get_image
from XAI.XAI_models import EIGEN_CAM
from XAI.BaseClass import get_model
from XAI.visualize import show_cam_on_image
```

```python
model = get_model('Temp/best.pt')
img, rgb_img = get_image('Temp/000023.jpg')
```

```python
target_layers =[model.model.model[-2]]
```

```python
cam = EIGEN_CAM(model, target_layers,task='od')
grayscale_cam = cam(rgb_img)[0, :, :]
show_cam_on_image(img, grayscale_cam, use_rgb=True)
# plt.imshow(cam_image)
# plt.show()
```
